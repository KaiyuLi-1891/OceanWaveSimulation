```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

x = np.linspace(-0, 30, 200)
y = np.linspace(-7.5, 7.5, 200)
X, Y = np.meshgrid(x, y)


fig, ax = plt.subplots()

ax.contour(X, Y, X, colors='lightgrey', linestyles='solid', linewidths=0.3, levels=np.arange(0, 30, 0.2))  # 绘制网格线
ax.contour(X, Y, Y, colors='lightgrey', linestyles='solid', linewidths=0.3, levels=np.arange(-7.5, 7.5, 0.2))  # 绘制网格线

inflow = Rectangle((0,-1.5),4,3,edgecolor='blue',facecolor='none',linewidth=2,zorder=3, hatch="|||")
sponge1 = Rectangle((0,7),30,0.5,edgecolor='red',facecolor='none',linewidth=1,zorder=3,hatch="///")
sponge2 = Rectangle((0,-7.5),30,0.5,edgecolor='red',facecolor='none',linewidth=1,zorder=3,hatch="///")
sponge3 = Rectangle((29.5,-7.5),0.5,15,edgecolor='red',facecolor='none',linewidth=1,zorder=3,hatch="///")

ax.axhline(y=1.5,linestyle='--',color='b',linewidth=1.5)
ax.axhline(y=-1.5,linestyle='--',color='b',linewidth=1.5)
    
ax.set_xlim(0, 30)
ax.set_ylim(-7.5, 7.5)
ax.set_yticks([-7.5, -6, -4.5, -3, -1.5, 0, 1.5, 3, 4.5, 6, 7.5])
ax.add_patch(inflow)
ax.add_patch(sponge1)
ax.add_patch(sponge2)
ax.add_patch(sponge3)
ax.set_xlabel('X')
ax.set_ylabel('Y',rotation=1800)
ax.set_aspect('equal')

plt.savefig("FlowDomain.png",dpi=700)
```
